```python
def calculate_pixels_from_top(img_data):
    pixels_from_top = []

    for column_index in range(0, 478):
        found_pixel = 0
        
        for row_index in range(0, 120):
            if img_data[row_index][column_index] == 0:
                pixels_from_top.append(row_index)
                found_pixel = 1
                break

        if found_pixel == 0:
            pixels_from_top.append(pixels_from_top[column_index - 1])

    return pixels_from_top
```

```python
def calculate_pixels_from_bottom(pixels_from_top):
    pixels_from_bottom = []

    for column_index in range(0, len(pixels_from_top)):
        pixels_from_bottom.append(120 - pixels_from_top[column_index])

    return pixels_from_bottom
```

```python
def plot_customized_signal(signal, j):
    
    
    
    plt.xticks(range(0,700,50))
    for i in range (0,len(signal),140):
        plt.axvline(x=i)
    plt.plot(signal, label='ECG Signal')
    np.save(f"output{j}", signal)

    plt.xlabel('X-Axis Label')
    plt.ylabel('Y-Axis Label')
    plt.legend()

    plt.show()
```

```python
def image_to_numpy_array(image_path):
    try:
        img = Image.open(image_path)

        img_gray = img.convert('L')

        img_data = np.array(img_gray)

        return img_data

    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
image_1=np.array(calculate_pixels_from_top(img_data_1))
image_1
```

```python
ecg_signal_1 = np.array(calculate_pixels_from_bottom(image_1))
ecg_signal_1
```

```python
plot_customized_signal(signal=ecg_signal_1, j=1)
```
